```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn.linear_model import LinearRegression
sns.set(style="ticks", color_codes=True)
%matplotlib inline
```

# Load and Clean Data

```python
# read the csv file
df = pd.read_csv('sentiment.csv')
print(df.shape)
```

```python
# delete null row
df.dropna(how='any',inplace=True)
print(df.shape)
```

```python
# find unusual rows
def set_as_nan(x):
    if x == 'time':
        return np.nan
    else:
        return x
df['time'] = df['time'].map(set_as_nan)
```

```python
# delete unusual rows
df.dropna(how='any',inplace=True)
print(df.shape)
```

```python
# select columns
df = df[['time','score','magnitude']]
```

```python
df['score'] = df['score'].map(lambda x:float(x))   
df['magnitude'] = df['magnitude'].map(lambda x:float(x)) # convert the type
df['sentiment'] = df['score'] * df['magnitude'] # get sentiment score 
```

```python
df = pd.DataFrame(df.groupby(['time'])['score','magnitude','sentiment'].mean()) #  get the mean value in the group of time
df = df.reset_index()
df = df.sort_values('time')
```

```python
df1 = pd.read_csv('hot.csv')
df2 = pd.read_csv('bitcoinprice.csv')
df1 = df1.sort_values(by='date')
df1.columns = ['time','googleview']
df2 = df2.sort_values(by='time')
df2.columns = ['time','price']
df1 = df1.reset_index()
df1 = df1[['time','googleview']]
df3 = pd.read_csv('transaction.csv')
df3 = df3.sort_values(by='Date')
df3.columns = ['time','volume']
df3 = df3.reset_index()
df3 = df3[['time','volume']]  # read the data which is stored in the csv
```

```python
df = df.merge(df1,how='inner',on='time')
df = df.merge(df2,how='inner',on='time')
df = df.merge(df3,how='inner',on='time') # merge dataframes
```

# Lagging Insight

```python
# lagged function
def lag(data,n=1):
    data = list(data)
    lagged_data = [data[i] for i in range(n,len(data))]
    for i in range(0,n):
        lagged_data.append(np.nan)
    return pd.Series(lagged_data) 
```

```python
gdf['sentiment_lag1'] = lag(gdf['sentiment'])
gdf['score_lag1'] = lag(gdf['score'])
gdf['magnitude_lag1'] = lag(gdf['magnitude'])
gdf['googleview_lag1'] = lag(gdf['googleview'])
gdf['volume_lag1'] = lag(gdf['volume'])
gdf['sentiment_lag2'] = lag(gdf['sentiment'],2)
gdf['score_lag2'] = lag(gdf['score'],2)
gdf['magnitude_lag2'] = lag(gdf['magnitude'],2)
gdf['googleview_lag2'] = lag(gdf['googleview'],2)
gdf['volume_lag2'] = lag(gdf['volume'],2)
gdf['sentiment_lag3'] = lag(gdf['sentiment'],3)
gdf['score_lag3'] = lag(gdf['score'],3)
gdf['magnitude_lag3'] = lag(gdf['magnitude'],3)
gdf['googleview_lag3'] = lag(gdf['googleview'],3)
gdf['volume_lag3'] = lag(gdf['volume'],3)
gdf['price_lag1'] = lag(gdf['price'])
gdf['price_lag2'] = lag(gdf['price'],2)
gdf['price_lag3'] = lag(gdf['price'],3)
gdf1=gdf # store the dataframe
# lag the columns 
```

# Multiple Variable Regression

```python
gdf=gdf.dropna() # drop the null data
for i in range(0,4): # since the bitcoin is exchanged in the finance market, we can see a high time series problem. so we try to fin the best combo of our factors.
    if i==0:
        reg=LinearRegression().fit(gdf[['sentiment','score','magnitude','googleview','volume','price']],gdf[['price_lag2']])
        print("Current Variable Regression")
        score = reg.score(gdf[['sentiment','score','magnitude','googleview','volume','price']],gdf[['price_lag2']])
    else:
        reg=LinearRegression().fit(gdf[['sentiment_lag'+str(i),'score_lag'+str(i),
                                        'magnitude_lag'+str(i),'googleview_lag'+str(i),
                                       'volume_lag'+str(i),'price_lag'+str(i)]],gdf[['price_lag2']])
        print("Lag "+str(i)+" Variable Regression")
        score = reg.score(gdf[['sentiment_lag'+str(i),
                       'score_lag'+str(i),      
                       'magnitude_lag'+str(i),
                       'googleview_lag'+str(i),
                              'volume_lag'+str(i),'price_lag'+str(i)]],gdf[['price_lag2']])
    print('Score: %.2f' % score )    
```

```python
# get the best solution 
reg=LinearRegression().fit(gdf[['sentiment','score','magnitude','googleview','volume','price_lag1']].iloc[:63],gdf[['price_lag2']].iloc[:63])
        
print("Current Variable Regression")
score = reg.score(gdf[['sentiment','score','magnitude','googleview','volume','price_lag1']].iloc[:63],gdf[['price_lag2']].iloc[:63])
print(score)
```

```python
prediction=reg.predict(gdf1[['sentiment','score','magnitude','googleview','volume','price_lag1']].iloc[63:64])
prediction=(round(prediction[0][0]*0.9, 2), round(prediction[0][0]*1.1,2))# test predicting the price 
print(prediction)
```
